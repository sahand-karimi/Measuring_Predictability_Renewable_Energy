```python
# The following REG FCAS costs are read from Quarterly energy dynamics q4 2020. Australian Energy Market Operator(AEMO); 2021.
# URL: https://aemo.com.au/-/media/files/majorpublications/qed/2020/qed-q4-2020.pdf

FCASshareWind = {
    'Q1': 21.818181818181813,
    'Q2': 13.553719008264459,
    'Q3': 14.545454545454545,
    'Q4': 19.338842975206607}

FCASshareSolar = {
    'Q1': 18.512396694214875,
    'Q2': 21.32231404958678,
    'Q3': 13.553719008264459,
    'Q4': 20.000000000000004}

FCASshareHydro = {
    'Q1': 6.942148760330575,
    'Q2': 7.93388429752066,
    'Q3': 6.611570247933884,
    'Q4': 6.280991735537182}

FCASshareBlackCoal = {
    'Q1': 8.264462809917351,
    'Q2': 6.776859504132225,
    'Q3': 6.611570247933884,
    'Q4': 3.140495867768586}

FCASshareGas = {
    'Q1': 5.454545454545447,
    'Q2': 1.8181818181818092,
    'Q3': 2.479338842975203,
    'Q4': 1.6528925619834685}

FCASshareBrownCoal = {
    'Q1': 0.6611570247933833,
    'Q2': 0.33057851239669167,
    'Q3': 0.9917355371900751,
    'Q4': 0.9917355371900751}
```

```python
# The following energy generation data of each fuel type is collcted from
# OpenNEM; URL: https://opennem.org.au/energy/nem/?range=all&interval=quarter.

WindTWh = {
    'Q1': 4.175,
    'Q2': 4.673,
    'Q3': 5.442,
    'Q4': 5.388}

SolarTWh = {
    'Q1': 1.647,
    'Q2': 1.287,
    'Q3': 1.520,
    'Q4': 2.248}

HydroTWh = {
    'Q1': 3.192,
    'Q2': 4.235,
    'Q3': 3.810,
    'Q4': 3.077}

BlackCoalTWh = {
    'Q1': 26.259,
    'Q2': 24.438,
    'Q3': 25.349,
    'Q4': 23.642}

GasTWh = {
    'Q1': 3.171,
    'Q2': 3.883,
    'Q3': 4.202,
    'Q4': 2.907}

BrownCoalTWh = {
    'Q1': 8.896,
    'Q2': 8.799,
    'Q3': 8.472,
    'Q4': 3.068}
```

```python
WindShareperTWh = {k: FCASshareWind[k]/WindTWh[k] for k in FCASshareWind.keys()}
SolarShareperTWh = {k: FCASshareSolar[k]/SolarTWh[k] for k in FCASshareSolar.keys()}
HydroShareperTWh = {k: FCASshareHydro[k]/HydroTWh[k] for k in FCASshareHydro.keys()}
GasShareperTWh = {k: FCASshareGas[k]/GasTWh[k] for k in FCASshareGas.keys()}
BlackCoalShareperTWh = {k: FCASshareBlackCoal[k]/BlackCoalTWh[k] for k in FCASshareBlackCoal.keys()}
BrownCoalShareperTWh = {k: FCASshareBrownCoal[k]/BrownCoalTWh[k] for k in FCASshareBrownCoal.keys()}
```

```python
# Plotting Figure 5

import matplotlib.pyplot as plt
import pandas as pd
from cycler import cycler

plt.style.use('default')
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['font.size'] = 7
plt.rcParams['figure.dpi'] = 400
plt.rcParams['savefig.dpi'] = 400

plt.rcParams['axes.prop_cycle'] = cycler('color', ['c', 'gold', 'forestgreen', 'dimgray', 'pink', 'brown'])

df = pd.DataFrame([WindShareperTWh ,SolarShareperTWh ,HydroShareperTWh , BlackCoalShareperTWh, GasShareperTWh, BrownCoalShareperTWh],
                  index=['Wind','Solar','Hydro','Black Coal', 'Gas', 'Brown Coal']).transpose()

fig, ax= plt.subplots(1,1, figsize=(3.35, 2))

df.plot.bar(ax=ax)
ax.set_xlabel('2020 Year Quarter')
ax.set_ylabel('Reg FCAS cost share per \n generated energy (%/TWh)', multialignment='center')

plt.legend(ncol=2, fancybox=True)

plt.grid(color = 'gainsboro', linestyle = '--', linewidth = 0.5, alpha = 0.5)
plt.gca().xaxis.grid(False)
plt.xticks(rotation=0)

plt.gca().set_yticklabels(['{:.1f}'.format(x) for x in plt.gca().get_yticks()])

plt.savefig('Figure5.pdf',dpi = 400, bbox_inches='tight')
```
